```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, pearsonr, spearmanr
import seaborn as sns
import matplotlib.pyplot as plt

# Generate dummy data
np.random.seed(42)
n = 50
usia = np.random.randint(17, 26, size=n)  # usia 17-25 tahun
persepsi_skor = np.random.randint(10, 21, size=n)  # skor persepsi antara 10-20

# Buat DataFrame
df = pd.DataFrame({
    'Usia': usia,
    'Persepsi_Keamanan': persepsi_skor
})

# Uji normalitas (Shapiro-Wilk)
stat, p_value_normalitas = shapiro(df['Persepsi_Keamanan'])

# Uji korelasi (tentukan berdasarkan distribusi)
if p_value_normalitas > 0.05:
    korelasi, p_value_korelasi = pearsonr(df['Usia'], df['Persepsi_Keamanan'])
    jenis_uji = "Pearson"
else:
    korelasi, p_value_korelasi = spearmanr(df['Usia'], df['Persepsi_Keamanan'])
    jenis_uji = "Spearman"

# Visualisasi korelasi
sns.set(style="whitegrid")
sns.regplot(x='Usia', y='Persepsi_Keamanan', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title(f'{jenis_uji} Correlation\nr = {korelasi:.2f}, p = {p_value_korelasi:.3f}')
plt.xlabel('Usia')
plt.ylabel('Skor Persepsi Keamanan')
plt.tight_layout()
plt.show()

# Ringkasan hasil
{
    "Jenis_Uji": jenis_uji,
    "Nilai_Korelasi": korelasi,
    "p_value": p_value_korelasi,
    "Normalitas_p_value": p_value_normalitas
}
```
